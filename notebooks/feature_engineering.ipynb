```python
import pandas as pd
df=pd.read_csv("../data/cybersecurity_cleaned.csv")
```

```python
#need dictionary
unique_protocol = df["protocol"].unique().tolist()
unique_protocol_List = []
for item in unique_protocol:
    unique_protocol_List.append(item)
unique_protocol_List
```

```python
protocol_dict = {}
for item in unique_protocol_List:
    protocol_dict[item] = unique_protocol_List.index(item)
protocol_dict
```

```python
#to retrieve the numbers
protocol_List = list(protocol_dict.values())
protocol_List
```

```python
unique_IsInternalTraffic = df["is_internal_traffic"].unique().tolist()
unique_IsInternalTraffic
```

```python
#not to be used in training
#retrieves all unique label values in the dataset
unique_label = df["label"].unique().tolist()
unique_label
```

```python
#not to be used in training
#retrieves all attack_type values in the dataset
unique_attack_type = df["attack_type"].unique().tolist()
unique_attack_type
```

```python
#assume works
df["timestamp"] = df["timestamp"].astype("datetime64[s]")
df
```

```python
#for src_ip and dst_ip
df["src_ip"] = df["src_ip"].str.replace('.','')
df["src_ip"] = df["src_ip"].astype(int)
df
```

```python
#for src_ip and dst_ip
df["dst_ip"] = df["dst_ip"].str.replace('.','')
df["dst_ip"] = df["dst_ip"].astype(int)
df
```

```python
df["is_internal_traffic"] = df["is_internal_traffic"].astype(int)
df
```

```python
df["protocol"] = df["protocol"].replace(protocol_dict)
df["protocol"] = df["protocol"].astype(int)
df
```
